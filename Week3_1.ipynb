#### The goal of this script is to scrape data of Toronto's neighborhoods and clean it in order to use in further analyses

First we import panda

```python
import pandas as pd # library for data analysis
```

Then we read the table from the wikipedia page (which is just first table)

```python
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
table=table[0]
```

Before deleting the rows where Borough is not assigned, let's check the values of neighborhoud where Borough is not assigned

```python
pd.set_option('display.max_rows', None)
table[table['Borough'] == 'Not assigned']
```

Neighborhoood is NaN at all times. We proceed by deleting these rows

```python
# Exclude rows where Borough is equal to 'Not assigned'
table = table[table['Borough'] != 'Not assigned']
```

Rename postal code to PostalCode as instructed

```python
#Rename postal code column
table.rename(columns={'Postal code': 'PostalCode'},inplace=True)
```

Let's look at the values for Neighborhood

```python
table[['Neighborhood']]
```

We see that in case of multiple neigborhoodss they are currently seperated with [space+forward slash + space]. 
So we need to replace this by with [comma + space]

```python
table['Neighborhood'] = table['Neighborhood'].str.replace(' / ',', ')
table
```

This looks like what we need. The final thing is to fix is to have consecutive row numbers

```python
table.reset_index(drop=True, inplace=True)
table
```
